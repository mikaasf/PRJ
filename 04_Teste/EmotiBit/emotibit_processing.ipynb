```python
filename = "ppg_signal.csv"

with open(filename) as ppg_file:
    ppg = csv.reader(ppg_file)
    
    data = []
    line_count = 0
    for row in ppg:
        if line_count:
            data.append(float(row[0]))
        
        line_count += 1
    
    data = np.array(data)
```

## Cálculo dos batimentos por minuto (BPM)

### Fórmula:

### $BPM = \frac{Sampling Rate}{(peak(i + 1) - peak(i))} \times 60 \text{       } (bpm)$

```python
# BPM = Sampling Rate / (peak(i + 1) - peak(i)) * 60

sampling_rate = 50

for i in range(1, peaks.size):
    if i + 1 != peaks.size:
        bpm = sampling_rate / (peaks[i + 1] - peaks[i]) * 60
        print("BPM:", int(np.round(bpm)))
```
